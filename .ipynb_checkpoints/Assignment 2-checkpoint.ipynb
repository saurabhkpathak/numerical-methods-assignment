```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fftpack import rfft, rfftfreq, irfft

def setup_graph(title='', x_label='', y_label='', fig_size=None):
    fig = plt.figure()
    if fig_size != None:
        fig.set_size_inches(fig_size[0], fig_size[1])
    ax = fig.add_subplot(111)
    ax.set_title(title)
    ax.set_xlabel(x_label)
    ax.set_ylabel(y_label)

t = np.linspace(-2, 2)
sample_rate = 100

def f1(x):
    global res
    if np.sqrt(x**2)>0.5:
        res = 0
    elif np.sqrt(x**2)==0.5:
        res = 1/2
    elif np.sqrt(x**2)<0.5:
        res = 1
    return res

vfunc = np.vectorize(f1)

setup_graph(x_label='time (in seconds)', y_label='amplitude', title='time domain')

plt.plot(t, vfunc(t))
plt.show()

fft_output = rfft(vfunc(t))

fft_magnitude_only = np.abs(fft_output) / len(fft_output)

frequencies_fft = rfftfreq(len(vfunc(t)),1/sample_rate) # Calculate all possible frequencies in signal.

setup_graph(x_label='frequency (in Hz)', y_label='amplitude', title='frequency domain')
plt.plot(frequencies_fft, fft_magnitude_only, 'r')
plt.show()
```

```python
def f2(x):
    global res2
    if x>0:
        res2 = 1
    elif x==0:
        res2 = 0
    elif x<0:
        res2 = -1
    return res2

vfunc2 = np.vectorize(f2)
setup_graph(x_label='time (in seconds)', y_label='amplitude', title='time domain')
plt.plot(t, vfunc2(t))
plt.show()

fft_output2 = rfft(vfunc2(t))

fft_magnitude_only2 = np.abs(fft_output2) / len(fft_output2)

frequencies_fft2 = rfftfreq(len(vfunc2(t)),1/sample_rate) # Calculate all possible frequencies in signal.

setup_graph(x_label='frequency (in Hz)', y_label='amplitude', title='frequency domain')
plt.plot(frequencies_fft2, fft_magnitude_only2, 'r')
plt.show()
```
